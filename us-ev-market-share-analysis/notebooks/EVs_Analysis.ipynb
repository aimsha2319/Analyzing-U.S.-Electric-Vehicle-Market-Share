```python
conversion_columns = ['Electric (EV)', 'Plug-In Hybrid Electric (PHEV)', 'Hybrid Electric (HEV)', 'Biodiesel', 'Ethanol/Flex (E85)', 'Compressed Natural Gas (CNG)', 'Propane',
                      'Hydrogen', 'Gasoline', 'Diesel', 'Unknown Fuel', 'Methanol']
```

```python
all_vehicles = ['Electric (EV)', 'Plug-In Hybrid Electric (PHEV)', 'Hybrid Electric (HEV)', 'Biodiesel', 
   'Ethanol/Flex (E85)', 'Compressed Natural Gas (CNG)', 'Propane','Hydrogen', 'Gasoline', 'Diesel', 'Unknown Fuel', 'Methanol']
```

```python
ev['Total_Vehicles'] = ev[['Electric (EV)', 'Plug-In Hybrid Electric (PHEV)', 'Hybrid Electric (HEV)', 'Biodiesel', 
   'Ethanol/Flex (E85)', 'Compressed Natural Gas (CNG)', 'Propane','Hydrogen', 'Gasoline', 'Diesel', 'Unknown Fuel', 'Methanol']].sum(axis=1)
```

```python
target_columns = ['Electric (EV)', 'Plug-In Hybrid Electric (PHEV)', 'Hybrid Electric (HEV)', 'Gasoline'] 
```

```python
percentages_calc = (ev[target_columns].div(ev['Total_Vehicles'], axis=0)*100).round(3)
```

```python
percentages_calc = percentages_calc.add_suffix('_perc')
```

```python
ev = pd.concat([ev, percentages_calc], axis=1)
```

```python
ev.sort_values(ascending=False, by="EVs_perc")
```
